Data Frame Creation Code

```python
import pandas as pd
import numpy as np

data = {
    'patient_id': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110,
                   111, 112, 113, 114, 115, 101, 107, 118, 119, 120],
    'age': ['25', '34', None, '45', '29', None, '38', '52', '27', '41',
            '33', 'unknown', '48', '26', '35', '25', '38', '31', None, '44'],
    'weight': ['70', '65', '80', None, '75', None, '68', '90', '72', '85',
               '78', None, '82', '69', 'N/A', '70', '68', '74', None, '88'],
    'blood_pressure': [120, 130, None, 140, 125, None, 135, None, 118, 145,
                      128, None, 138, 122, None, 120, 135, 126, None, 142],
    'medication': ['Aspirin', 'Metformin', 'Lisinopril', None, 'Aspirin',
                   'Metformin', 'Lisinopril', 'Aspirin', None, 'Metformin',
                   'Lisinopril', 'Aspirin', None, 'Metformin', 'Aspirin',
                   'Aspirin', 'Lisinopril', 'Metformin', 'Aspirin', None],
    'insurance_provider': ['Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', None,
                          'Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', 'Blue Cross',
                          'Aetna', None, 'UnitedHealth', 'Blue Cross', 'Aetna',
                          'Blue Cross', 'Aetna', 'Cigna', 'UnitedHealth', None]
}

df = pd.DataFrame(data)
print(df)
```

**Task 1: Inspect the Data**

```python
# 1.Use df.info() to view column data types and non-null counts
print(df.info())
```

```python
#2. Use df.isnull().sum() to count missing values per column
print(df.isnull().sum())
```

```python
#3. Calculate the percentage of missing values using: (df.isnull().sum() / len(df)) * 100
print((df.isnull().sum() / len(df)) * 100)
```

```python
#4. Use df.duplicated().sum() to count duplicate rows
print(df.duplicated().sum())
```

**Task 2: Data Type Conversion**

```python
#1.Convert age column to numeric using pd.to_numeric(df['age'], errors='coerce')
df_cleaned = df.copy()
df_cleaned['age'] = pd.to_numeric(df_cleaned['age'],errors='coerce')
```

```python
#2. Convert weight column to numeric using pd.to_numeric(df['weight'], errors='coerce')
df_cleaned['weight'] = pd.to_numeric(df_cleaned['weight'],errors='coerce')
```

```python
#3. Check how many NEW missing values were created after conversion
# Age and weight were converted to numeric using errors='coerce'.
# Any non-numeric values like 'unknown' (age) and 'N/A' (weight) were converted into NaN.
# Because of this, the number of missing values increased.
print("---------------Printing count of Null elements column wise in ORIGINAL DF---------------")
print(df.isnull().sum())
print("---------------Printing count of Null elements column wise in ALTERED DF---------------")
print(df_cleaned.isnull().sum())
print("---------------Printing difference of Null elements column---------------")
print((df_cleaned.isnull().sum()) - df.isnull().sum())
```

```python
#4. Convert insurance_provider to category type using astype('category')
df_cleaned['insurance_provider'] = df_cleaned['insurance_provider'].astype('category')

#Printing the catergory wise count
print(df_cleaned['insurance_provider'].value_counts(dropna=False))
```

```python
#5. Use df.dtypes to verify all conversions
print(df_cleaned.dtypes)
```

**Task 3: Handle Missing Values**

```python
df_handle_missing = df_cleaned.copy()
# 1. age: Fill with median using fillna(df['age'].median())
df_handle_missing['age'] = df_handle_missing['age'].fillna(df_handle_missing['age'].median())

# 2. weight: Fill with median using fillna(df['weight'].median())
df_handle_missing['weight'] = df_handle_missing['weight'].fillna(df_handle_missing['weight'].median())

# 3. blood_pressure: Fill with median using fillna(df['blood_pressure'].median())
df_handle_missing['blood_pressure'] = df_handle_missing['blood_pressure'].fillna(df_handle_missing['blood_pressure'].median())

# 4. medication: Fill with mode using fillna(df['medication'].mode()[0])
df_handle_missing['medication'] = df_handle_missing['medication'].fillna(df_handle_missing['medication'].mode()[0])

# 5. insurance_provider: Fill with constant value 'Unknown'
#Since, insurance provider is already a category, we need to add Unknown as a new category first before filling the missing values as Unknown
df_handle_missing['insurance_provider'] = df_handle_missing['insurance_provider'].cat.add_categories('Unknown')
#Now Handle missing values further
df_handle_missing['insurance_provider'] = df_handle_missing['insurance_provider'].fillna('Unknown')

print(df_handle_missing)
```

**Task 4: Handle Duplicates**

```python
df_handle_duplicates = df_handle_missing.copy()
#1. View duplicate rows using df[df.duplicated(keep=False)]
print(df_handle_duplicates[df_handle_duplicates.duplicated(keep=False)])
#2. Identify duplicates based on patient_id only using df.duplicated(subset=['patient_id'])
print(df_handle_duplicates[df_handle_duplicates.duplicated(subset=['patient_id'])])
#3. Remove duplicates keeping first occurrence: df.drop_duplicates(subset=['patient_id'], keep='first')
df_handle_duplicates = df_handle_duplicates.drop_duplicates(subset='patient_id',keep='first')
print(df_handle_duplicates)

#4. Print shape before and after to show how many rows were removed
print(f"Shape Before : {df_handle_missing.shape}")
print(f"Shape After : {df_handle_duplicates.shape}")

```

**Task 5: Complete Workflow with Verification**

```python
import pandas as pd
import numpy as np

# ---------------------------
# 1) Start Fresh
# ---------------------------
df_clean = df.copy()

print("="*50)
print("\t\tBEFORE CLEANING")
print("="*50)

print(f"Shape: {df_clean.shape}\n")

print("Missing Values:")
print(df_clean.isnull().sum())
print()

print(f"Duplicate Rows: {df_clean.duplicated().sum()}\n")

print("Data Types:")
print(df_clean.dtypes)


# ---------------------------
# 2) TYPE CONVERSION
# ---------------------------
df_clean['age'] = pd.to_numeric(df_clean['age'], errors='coerce')
df_clean['weight'] = pd.to_numeric(df_clean['weight'], errors='coerce')


# ---------------------------
# 3) HANDLE MISSING VALUES
# ---------------------------
df_clean['age'] = df_clean['age'].fillna(df_clean['age'].median())
df_clean['weight'] = df_clean['weight'].fillna(df_clean['weight'].median())
df_clean['blood_pressure'] = df_clean['blood_pressure'].fillna(df_clean['blood_pressure'].median())
df_clean['medication'] = df_clean['medication'].fillna(df_clean['medication'].mode()[0])
df_clean['insurance_provider'] = df_clean['insurance_provider'].fillna('Unknown')

df_clean['insurance_provider'] = df_clean['insurance_provider'].astype('category')


# ---------------------------
# 4) REMOVE DUPLICATES
# ---------------------------
df_clean = df_clean.drop_duplicates(subset=['patient_id'], keep='first')


# ---------------------------
# 5) VERIFICATION REPORT
# ---------------------------
print("\n" + "="*50)
print("\t\tAFTER CLEANING")
print("="*50)

print(f"Shape: {df_clean.shape}\n")

print("Missing Values:")
print(df_clean.isnull().sum())
print()

print(f"Duplicate Rows: {df_clean.duplicated().sum()}\n")

print("Data Types:")
print(df_clean.dtypes)
```
